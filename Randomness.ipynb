 Several years ago, there was a documentary on Discovery Channel about a Russian girl who claims that she can find the previously injured body parts of the people by just looking at them. The local people were convinced about the girl's talent. It was amazing because she just looks at the people and says "you had an operation or have a problem here" etc. The producer of the document also challenges her talent and he is also shocked that she could find an old injury he had some time ago. So he takes her to US to show scientists and ask them to test her talent. The scientists set up an experiment where there are seven people with known medical history. They write down the medical history of each person and shuffle the cards. Then they ask the girl to match the cards with the correct people. If she can correctly match 5 of them, they will think that she has really some talent in this issue. Note that she cannot have 6 matches because one mistake will also cause another mistake. However, she could correctly match only 4 of them. So they think that her talent is bogus and do not believe her. 
 What is the probability of having 4 matches out of 7 if the cards are distributed randomly? Find the probability by simulation using 10000 trials.

```python
# write your answer here
import random
import numpy as np

diagnosis= make_array('d1', 'd2', 'd3','d4', 'd5', 'd6','d7')
people = make_array('d1', 'd2', 'd3','d4', 'd5', 'd6','d7')


def finding(obj,array):
    for i in range(0,array.size):
        if obj == array[i]:
            return i
        else:
            continue
    return -1



def method():
    dummy_diagnosis=np.copy(diagnosis)
    dummy_people=np.copy(people)
    counter=0
    while 5>1:
            if dummy_people.size ==0:
                return counter
            original_diagnose = np.random.choice(dummy_diagnosis)
            original_person = np.random.choice(dummy_people)
            if original_diagnose ==original_person:
                counter+=1
            index_d=finding(original_diagnose,dummy_diagnosis)
            index_p=finding(original_person,dummy_people)
            dummy_diagnosis= np.delete(dummy_diagnosis,index_d)
            dummy_people=np.delete( dummy_people,index_p)
    print("x")
    return counter


def problem(i,N):
    counter = 0
    for i in range (1,i+1):
         if method() == N:
            counter+=1
    return counter/i 



```

 Change the number of trials from 1K, 2K, 4K, 8K, 16K, 32K, 64K, and 128K and visualize your results where x axis is the number of trials and y axis is the probability of having 4 matches.

```python
# write your answer here
x_axes = make_array("1K", "2K", "4K","8K", "16K", "32K", "64K")
y_axes =make_array(problem(1000,4),problem(2000,4),problem(4000,4),problem(8000,4),problem(16000,4),problem(32000,4),problem(64000,4))

results = Table().with_columns(
    'X Axes', x_axes,
    'Y Axes', y_axes
    
).plot("X Axes")










```
